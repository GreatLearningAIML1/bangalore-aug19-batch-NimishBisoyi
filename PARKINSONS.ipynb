4. Split the dataset into training and test set in the ratio of 70:30 (Training:Test)

```python
data = data.drop("name" , axis = 1)
```

```python
X = data.drop("status" , axis = 1)
Y = data[["status"]]
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=7)

```

```python
print('x_train',x_train.shape)
print('x_test',x_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```

5. Create the model using â€œentropyâ€ method of reducing the entropy and fit it to training data.

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(criterion = 'entropy',random_state = 100 )
```

```python
dt_model.fit(x_train, y_train)
```

```python
from IPython.display import Image  
#import pydotplus as pydot
from sklearn import tree
from os import system

train_char_label = ['No', 'Yes']

Parkinson_Tree_File = open('parkinson_tree.dot','w')
dot_data = tree.export_graphviz(dt_model, out_file=Parkinson_Tree_File, feature_names = list(x_train), class_names = list(train_char_label))

Parkinson_Tree_File.close()

# importance of features in the tree building ( The importance of a feature is computed as the 
#(normalized) total reduction of the criterion brought by that feature. It is also known as the Gini importance )

print (pd.DataFrame(dt_model.feature_importances_, columns = ["Imp"], index = x_train.columns))
```

```python
dt_model.score(x_test , y_test)
```

```python
system("dot -Tpng parkinson_tree.dot -o parkinson_tree.png")
Image("parkinson_tree.png")
```

6. Test the model on test data and what is the accuracy achieved. Capture the predicted values and do a crosstab

```python
y_predictTrain = dt_model.predict(x_train)
```

```python
y_predictTest = dt_model.predict(x_test)
```

```python
accuracy_score(y_test,y_predictTest)
```

```python
pd.crosstab(y_train,pd.Series(y_predictTrain),rownames=['Actual'],colnames=['Prediction'])
```

7.Use regularization parameters of max_depth, min_sample_leaf to recreate the model. What is the impact on the model accuracy? How does regularization help?

```python
reg_dt_model = DecisionTreeClassifier(criterion = 'entropy')
reg_dt_model.fit(x_train, y_train)
y_predictTest =reg_dt_model.predict(x_test)
```

```python
accuracy_score(y_test,y_predictTest)
```

```python
reg_dt_model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3,min_samples_leaf=5)
reg_dt_model.fit(x_train, y_train)
y_predictTest =reg_dt_model.predict(x_test)
```

```python
accuracy_score(y_test,y_predictTest)
```
