### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
genderData = hr_Data['Gender']
```

```python
maleCount = (genderData==1).value_counts().loc[True]      # assuming 1 is Male 
femaleCount = (genderData==2).value_counts().loc[True]    # assuming 2 is Female

print("Number of Male employees   : ", maleCount)
print("Number of Female employees : ", femaleCount)
```

### Find out the mean, median and variance for the column - DailyRate 

```python
dailyRateData = hr_Data['DailyRate']

medianVal = dailyRateData.median()
meanVal = dailyRateData.mean()
varianceValue = dailyRateData.var()

# print results
print('Mean     :', meanVal)
print('Median   :', medianVal)
print('Variance :', varianceValue)
```

### Plot the attrition column (Since it is a categorical variable, chose the plot accordingly)

```python
%matplotlib inline
import matplotlib.pyplot as plt

attr_counter = {} # dictionary
for z in hr_Data.itertuples():    
    attr_counter[z[2]] = attr_counter.get(z[2], 0) + 1        
```

```python
plt.pie(attr_counter.values(), labels=attr_counter.keys())
plt.title('Attrition Distribution')
plt.show()
```
